```python
MENU_INICIAL = (
    "Digite W para escrever, R para ler, "
    "L para listar toda a memoria e qualquer tecla para parar."
)
MENU_CURTO = "Digite W para escrever, R para ler e qualquer tecla para parar."


# ──────────────────────────────────────────────
# Utilitários
# ──────────────────────────────────────────────

def validar_bits(valor: str, n: int) -> bool:
    """Verifica se 'valor' é uma string binária de exatamente n bits."""
    return len(valor) == n and all(c in '01' for c in valor)


# ──────────────────────────────────────────────
# Operações de memória
# ──────────────────────────────────────────────

def listar_memoria(memoria: list) -> None:
    print("\nTodos os estados:")
    for dado in memoria:
        print(dado)
    print()


def escrever(memoria: list) -> None:
    endereco = input("Digite o endereco de 4 bits: ").strip()
    if not validar_bits(endereco, 4):
        print("Erro: o endereco deve ter exatamente 4 bits (ex: 0000, 1010)\n")
        return

    dado = input("Digite o dado de 8 bits: ").strip()
    if not validar_bits(dado, 8):
        print("Erro: o dado deve ter exatamente 8 bits (ex: 10110001)\n")
        return

    indice = int(endereco, 2)
    memoria[indice] = dado
    print()


def ler(memoria: list) -> None:
    endereco = input("Digite o endereco: ").strip()
    if not validar_bits(endereco, 4):
        print("Erro: o endereco deve ter exatamente 4 bits (ex: 0000, 1010)\n")
        return

    indice = int(endereco, 2)
    print(memoria[indice])
    print()


# ──────────────────────────────────────────────
# Loop principal
# ──────────────────────────────────────────────

def main() -> None:
    memoria = ['00000000'] * 16          # 16 posições inicialmente zeradas
    print(MENU_INICIAL)

    while True:
        cmd = input().strip().upper()

        if cmd == 'W':
            escrever(memoria)
        elif cmd == 'R':
            ler(memoria)
        elif cmd == 'L':
            listar_memoria(memoria)
        else:
            print("Encerrando simulador.")
            break

        print(MENU_CURTO)


if __name__ == "__main__":
    main()
```
